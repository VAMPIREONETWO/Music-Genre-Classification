```python
import librosa
from librosa import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
sr =16*1000
x, sr = librosa.load("Data/genres_original/jazz/jazz.00053.wav", sr=sr, mono=True)
librosa.display.waveshow(x,sr=sr, color="blue")
```

```python
x = librosa.util.normalize(x)
librosa.display.waveshow(x,sr=sr, color="blue")
```

```python
mel_spect = librosa.feature.melspectrogram(y=x,sr =sr)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
```

```python
librosa.display.specshow(mel_spect, sr=sr)
plt.show()
```
